```python
import pandas as pd
import matplotlib as plt

%matplotlib inline
%config InlineBackend.figure_format ='retina'
```

```python
rookie20 = pd.read_csv('../Project/Rookie20.csv',delimiter=",")
```

```python
rookie20['PerGameCompScore'] = (rookie20['MP.1']+rookie20['PTS.1']+rookie20['TRB.1']+rookie20['AST.1'])/4
R20 = rookie20.sort_values(by='PerGameCompScore',ascending=False)
```

```python
R20['Name'] = R20['Player'].apply(lambda x: x.split('\\')[0])
```

```python
Sub = R20[(R20['PerGameCompScore']>=6)]
```

```python
NBAplot = Sub[['Name','PerGameCompScore']]
```

```python
nplot = NBAplot.plot(kind='bar',figsize=(15,7), color=('orange'), xlabel="Rookie Name", ylabel="Composite Score", title='Composite Scores of 2020 NBA Rookies (With Scores Greater Than 6)')
ha = ['right', 'center', 'left']
nplot.set_xticklabels(nplot.get_xticks(), rotation = 45, ha=ha[0])
nplot.set_xticklabels(Sub['Name'])
nplot;

```

```python
nplot.figure.savefig('assignment 4 plot.pdf', bbox_inches='tight')
```
