>Using the data below, answer the following questions:

>1) Which entities (top 5) had the largest population density in 2020?
>2) Which entities have more water area than land area?
>3) Which entities increased in population the most in the last 10 years?
>4) What state bird accounts for the largest population as of 2020? Land area?
>5) How many entities' largest city is their capital city?
>6) Which entity has the largest percent drop from their largest city to their 5th largest? 100*(1st largest - 5th largest)/(1st largest)

```python
import pandas as pd
```

```python
facts = pd.read_csv('../data/state_facts.tsv',delimiter="\t")
facts.head(5)
```

```python
# Question 1 
facts['PopDensity2020'] = facts['Pop_2020']/facts['Area_tot'] 
facts = facts.sort_values(by='PopDensity2020',ascending=False)
facts.head(5)
```

```python
# Question 2 
facts[(facts['Area_water']>facts['Area_land'])]
```

```python
# Question 3 
facts['PopDifference'] = facts['Pop_2020']-facts['Pop_2010'] 
facts = facts.sort_values(by='PopDifference',ascending=False)
facts.head(3)
```

```python
# Question 4
bird1 = facts[['Pop_2020','State_bird']].groupby(by='State_bird').sum()
bird1 = bird1.sort_values(by='Pop_2020',ascending=False)
bird1.head(3)
```

```python
bird2 = facts[['Area_land','State_bird']].groupby(by='State_bird').sum()
bird2 = bird2.sort_values(by='Area_land',ascending=False)
bird2.head(3)
```

```python
# Question 5
Cap = facts[(facts['Capital_pop']>facts['city_1_pop'])&(facts['Capital_pop']>facts['city_2_pop'])&(facts['Capital_pop']>facts['city_3_pop'])&(facts['Capital_pop']>facts['city_4_pop'])&(facts['Capital_pop']>facts['city_5_pop'])]
NumberCities = len(Cap)
print('There are', NumberCities, 'cities where the largest city is the capital city')

```

```python
# Question 6
facts['percentDrop'] = 100*(facts['city_1_pop']-facts['city_5_pop'])/facts['city_1_pop']
facts[['State','City_1','city_1_pop','City_5','city_5_pop','percentDrop']].sort_values(by='percentDrop', ascending=False).head()
```

```python
shortFacts = pd.read_csv('../data/state_dates.tsv',delimiter="\t")
shortFacts
```

```python
joinedFacts = pd.merge(facts,shortFacts,left_on='USPS_code',right_on='Abbreviation',how='outer')
joinedFacts
```

```python
#Question 7
joinedFacts['Date_Fixed'] = pd.to_datetime(joinedFacts['Date'],errors='raise')
#joinedFacts[['Date','Date_Fixed']]
joinedFacts['Before1970'] = joinedFacts['Date_Fixed']<pd.datetime(1790,1,1)
Filtered = joinedFacts[(joinedFacts['Before1970']==True)]
Filtered = Filtered.groupby(by='State_flower').count()
print(Filtered['State'])
print("The Mountain laurel is the most common flower amongst the states established before 1970")
```
