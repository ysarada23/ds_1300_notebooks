Rookie Analysis Scratch Work

```python
import pandas as pd
```

```python
R21 = pd.read_csv('../Project/Rookie21.csv',delimiter=",")
Adv21 = pd.read_csv('../Project/Advanced21.csv',delimiter=",")
R21['UID'] = R21['Player'].apply(lambda x: x.split('\\')[1])
Adv21['UID'] = Adv21['Player'].apply(lambda x: x.split('\\')[1])
Radv21 = pd.merge(R21,Adv21, left_on="UID", right_on="UID", how = 'inner')
RF21 = Radv21.drop(columns=['Unnamed: 19', 'Unnamed: 24'])

RF21['PerGameCompScore'] = (RF21['MP.1']+RF21['PTS.1']+RF21['TRB.1']+RF21['AST.1'])/2
RF21['Stat%'] = (RF21['TRB%']+RF21['AST%']+RF21['STL%']+RF21['BLK%']+RF21['TOV%']+RF21['USG%'])*(RF21['VORP']+RF21['VORP'])
RF21['Combo'] = (RF21['Stat%']+RF21['PerGameCompScore'])/2
RF21['Hold'] = (RF21['PER']*RF21['TS%'])*RF21['WS']
RF21['Composite'] = RF21['Hold'] + (RF21['PTS']/10) + RF21['Combo'] + (RF21['MP_x']/10) 
RF21 = RF21.sort_values(by='Composite',ascending=False)
RF21.head(3)
```

```python
import matplotlib as plt
%matplotlib inline
%config InlineBackend.figure_format ='retina'
RF21['Name'] = RF21['Player_x'].apply(lambda x: x.split('\\')[0])
RF21['Likelihood of Winning Rookie of the Year'] = (RF21['Composite']/400)*100
Sub = RF21[(RF21['Likelihood of Winning Rookie of the Year']>=38)]
NBAplot = Sub[['Name','Likelihood of Winning Rookie of the Year']]
nplot = NBAplot.plot(kind='bar',figsize=(15,7), color=('orange'), legend='hello', xlabel="Rookie Name", ylabel="Scaled Composite Score", title='2021 NBA Rookie of the Year Odds')
ha = ['right', 'center', 'left']
nplot.set_xticklabels(nplot.get_xticks(), rotation = 45, ha=ha[0])
nplot.set_xticklabels(Sub['Name'])
nplot;
```

```python
nplot.figure.savefig('2021 Predications.pdf', bbox_inches='tight')
```

```python
rookie20 = pd.read_csv('../Project/Rookie20.csv',delimiter=",")
rookie20['PerGameCompScore'] = (rookie20['MP.1']+rookie20['PTS.1']+rookie20['TRB.1']+rookie20['AST.1'])/2
R20 = rookie20.sort_values(by='PerGameCompScore',ascending=False)

Adv20 = pd.read_csv('../Project/Advanced20.csv',delimiter=",")
R20['UID'] = R20['Player'].apply(lambda x: x.split('\\')[1])
Adv20['UID'] = Adv20['Player'].apply(lambda x: x.split('\\')[1])
Radv20 = pd.merge(R20,Adv20, left_on="UID", right_on="UID", how = 'inner')
RF20 = Radv20.drop(columns=['Unnamed: 19', 'Unnamed: 24'])


RF20['Stat%'] = (RF20['TRB%']+RF20['AST%']+RF20['STL%']+RF20['BLK%']+RF20['TOV%']+RF20['USG%'])*(RF20['VORP']+RF20['VORP'])
RF20['Combo'] = (RF20['Stat%']+RF20['PerGameCompScore'])/2
RF20['Hold'] = (RF20['PER']*RF20['TS%'])*RF20['WS']
RF20['Composite'] = RF20['Hold'] + (RF20['PTS']/10) + RF20['Combo'] + (RF20['MP_x']/10) 
RF20 = RF20.sort_values(by='Composite',ascending=False)
RF20.head(1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 2, figsize=(15, 10), constrained_layout=True)
fig.suptitle('Various NBA Statistics And Their Correlation With The Developed Composite Score Algorithm', fontsize=15)

axs[0, 0].scatter(RF20['Composite'],RF20['PTS'], color='orange')
axs[0, 0].set_title('Points Scored (Entire Season) vs. Composite Score', fontsize=12)
axs[0, 0].set_xlabel('Composite Score')
axs[0, 0].set_ylabel('Points')
axs[0, 0].grid(True, linestyle='-.')

axs[1, 0].scatter(RF20['Composite'],RF20['Stat%'], color='mediumseagreen')
axs[1, 0].set_title('Stat% (Accumulated Player Performance Stats) vs. Composite Score', fontsize=12)
axs[1, 0].set_xlabel('Composite Score')
axs[1, 0].set_ylabel('Stat%')
axs[1, 0].grid(True, linestyle='-.')

axs[0, 1].scatter(RF20['Composite'],RF20['MP_x'], color='royalblue')
axs[0, 1].set_title('Minutes Played (Entire Season) vs. Composite Score', fontsize=12)
axs[0, 1].set_xlabel('Composite Score')
axs[0, 1].set_ylabel('Minutes Played')
axs[0, 1].grid(True, linestyle='-.')

axs[1, 1].scatter(RF20['Composite'],RF20['Hold'], color='crimson')
axs[1, 1].set_title('Hold (Accumulated Player Value Stats) vs. Composite Score', fontsize=12)
axs[1, 1].set_xlabel('Composite Score')
axs[1, 1].set_ylabel('Hold')
axs[1, 1].grid(True, linestyle='-.')


plt.show()
```

```python
fig.figure.savefig('Composite Correlations.pdf', bbox_inches='tight')
```
